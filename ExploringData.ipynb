## Cleaning data
We'll eliminate all the rows with at least one missing value, and save the new matrix tx in clean_data, with the corresponding signals in clean_signals


```python
clean_data = []
clean_signals = []
for i,row in enumerate(tx):
    count= 0
    for el in row: 
        if el == -999.0:
            count+=1
    if count == 0:
        clean_data.append(row)
        clean_signals.append(signals[i])
clean_data=np.array(clean_data)
clean_signals=np.array(clean_signals)
```

# GRADIENT DESCENT

```python
def compute_loss(y, tx, w):
    e = y-tx.dot(w)
    return (e.dot(e))/2*e.size  # MSE
```

```python
def compute_gradient(y, tx, w):
    e = y-tx.dot(w)
    return -(tx.transpose().dot(e))/e.size
```

```python
def leastsquaresGD(y, tx, initial_w, max_iters, gamma):
    ws = [initial_w]
    losses = []
    w = initial_w
    for n_iter in range(max_iters):
        grad = compute_gradient(y, tx, w)
        loss = compute_loss(y, tx, w)
        w = w - gamma*grad
        ws.append(w)
        losses.append(loss)
        
        #---print only the first 4 weigths to have an idea of what's going on----
        
        print("Gradient Descent({bi}/{ti}): loss={l}, w0={w0}, w1={w1}, w2={w2}, w3={w3}, w4={w4}".format(
              bi=n_iter, ti=max_iters - 1, l=loss, w0=w[0], w1=w[1], w2=w[2], w3=w[3], w4=w[4]))

    return losses, ws
```
